```python
import tensorflow as tf
import numpy as np
import re

# Download and Load text
path_to_file = tf.keras.utils.get_file('alice.txt', 'http://www.gutenberg.org/files/11/11-0.txt')
with open(path_to_file, "r", encoding="utf-8") as f:
    text = f.read().lower()

# Clean text
text = re.sub(r'[^a-z\s]', '', text)

# Tokenization
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts([text])
sequence = tokenizer.texts_to_sequences([text])[0]
vocab_size = len(tokenizer.word_index) + 1
```

```python
input_sequences = []
sequence_length = 20

for i in range(sequence_length, len(sequence)):
    input_sequences.append(sequence[i-sequence_length:i+1])

input_sequences = np.array(input_sequences)

X = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 64, input_length=sequence_length),
    tf.keras.layers.LSTM(128, return_sequences=True),
    tf.keras.layers.LSTM(128),
    tf.keras.layers.Dense(vocab_size, activation='softmax')
])

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
history = model.fit(X, y, epochs=20, batch_size=128)
```

```python
def generate_text(seed_text, next_words=30):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = token_list[-sequence_length:]
        token_list = tf.keras.preprocessing.sequence.pad_sequences(
            [token_list], maxlen=sequence_length, padding="pre"
        )
        predicted = np.argmax(model.predict(token_list))
        output_word = tokenizer.index_word[predicted]
        seed_text += " " + output_word
    return seed_text
```
